```python
import pygame
import tkinter as tk
from tkinter import filedialog as fd
from tkinter import *
from tkinter.messagebox import showinfo

root = tk.Tk()
root.title("Mock 1 - Play Pause Test")
root.geometry("450x200")

pygame.init()
pygame.mixer.init()

var = ""
text = tk.StringVar()

def PLAY():
    if var != "":
        pygame.mixer.music.set_volume(1.0)
        pygame.mixer.music.load(var)
        pygame.mixer.music.play()
    else:
        showinfo(
            title="Invalid!",
            message="No file selected or invalid format!"
        )
        

def PAUSE():
    if pygame.mixer.music.get_busy():
        pygame.mixer.music.pause()
    else: 
        pygame.mixer.music.unpause()
            
lbl = tk.Label(
    root, 
    font=("Arial", 15),
    textvariable=text
).place(x=125, y=10)    


play = tk.Button(
    root,
    height=1,
    text="PLAY",
    command=PLAY
).place(x=50, y=50)

pause = tk.Button(
    root,
    height=1,
    text="PAUSE",
    command=PAUSE
).place(x=100, y=50)

Next = tk.Button(
    root,
    height=1,
    text="NEXT",
    command=None
).place(x=155, y=50)

Prev = tk.Button(
    root,
    height=1,
    text="PREV",
    command=None
).place(x=205, y=50)

def Stop():
    pygame.mixer.music.stop()
    pygame.mixer.music.unload()
    
stop = tk.Button(
    root,
    height=1,
    text="STOP",
    command=Stop
).place(x=50, y=95)

def Rewind():
    pygame.mixer.music.rewind()

rewind = tk.Button(
    root,
    height=1,
    text="REWIND",
    command=Rewind
).place(x=100, y=95)


def OpenFile():
    global var
    
    filetypes = (
        ('mp3 (*.mp3)', '*.mp3'),
        ('wav (*.wav)', '*.wav'),
        ('ogg (*.ogg)', '*.ogg'),
        ('wma (*.wma)', '*wma'),
        ('m4a (*.m4a)', '*.m4a'),
        ('flac (*.flac)', '*.flac'),
        ('aac (*.aac)', '*.aac'),
        ('All Files (*.*)', '*.*')
    )
    
    var = fd.askopenfilename(
        title = "open a file",
        initialdir = "/",
        filetypes=filetypes
    )
    
    y = var.split("/")[-1]
    z = y.split(".")[0]
    text.set(z)

Open = tk.Button(
    root,
    height=1,
    text="OPEN",
    command=OpenFile
).place(x=50, y=10)

def setVolume(vol):
    pygame.mixer.music.set_volume(int(vol)/100)

slid_var = DoubleVar()
slid_var.set(100)
slider = Scale(
    root,
    from_=0, to=100, 
    orient=HORIZONTAL,
    command=setVolume,
    variable=slid_var
).place(x=15, y=150)


root.mainloop()
```
